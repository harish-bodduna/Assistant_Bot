```python
def is_logo(image_path, reference_hash, threshold=11):
    """
    Compares an image against the reference hash.
    threshold: 0 is an exact match. 5-10 allows for slight scaling/artifacts.
    """
    test_image = Image.open(image_path)
    test_hash = imagehash.phash(test_image)
    
    # Calculate Hamming distance (number of bits that differ)
    distance = test_hash - reference_hash
    
    return distance < threshold, distance
```

```python
images_to_keep = []
images_to_discard = []

root = Path(
    "C:/Users/Harish/Workspace/maestro_projects/1440_Bot/"
    "markdown_exports/External_SharePoint_access/images"
)

for img_file in sorted(os.listdir(root)):
    path = os.path.join(root, img_file)
    
    match, score = is_logo(path, logo_hash)
    
    if match:
        print(f"[-] DISCARDING: {img_file} (Distance: {score})")
        images_to_discard.append(img_file)
    else:
        print(f"[+] KEEPING: {img_file} (Distance: {score})")
        images_to_keep.append(img_file)
```
